```python
# Calculate weighted mean precipitation isotope value for each season across each watershed
# And calculate isotope standard errors
# Then compile data in a table to export to csv
import pandas as pd
import math

table = r"C:\Users\User\Documents\UNR\NEON Project\Data\GIS\WaterIsotopesDatabase\tables.gdb\sum_ppt_summer"
columns = [f.name for f in arcpy.ListFields(table) if f.type!="Geometry"] 
sum_ppt_summer = pd.DataFrame(data=arcpy.da.SearchCursor(table, columns), columns=columns)

table = r"C:\Users\User\Documents\UNR\NEON Project\Data\GIS\WaterIsotopesDatabase\tables.gdb\sum_ppt_winter"
columns = [f.name for f in arcpy.ListFields(table) if f.type!="Geometry"] 
sum_ppt_winter = pd.DataFrame(data=arcpy.da.SearchCursor(table, columns), columns=columns)

table = r"C:\Users\User\Documents\UNR\NEON Project\Data\GIS\WaterIsotopesDatabase\tables.gdb\sum_ppt_x_d2h_summer"
columns = [f.name for f in arcpy.ListFields(table) if f.type!="Geometry"] 
sum_ppt_x_d2h_summer = pd.DataFrame(data=arcpy.da.SearchCursor(table, columns), columns=columns)

table = r"C:\Users\User\Documents\UNR\NEON Project\Data\GIS\WaterIsotopesDatabase\tables.gdb\sum_ppt_x_d2h_winter"
columns = [f.name for f in arcpy.ListFields(table) if f.type!="Geometry"] 
sum_ppt_x_d2h_winter = pd.DataFrame(data=arcpy.da.SearchCursor(table, columns), columns=columns)

table = r"C:\Users\User\Documents\UNR\NEON Project\Data\GIS\WaterIsotopesDatabase\tables.gdb\summer_d2h_se_sq_wtd"
columns = [f.name for f in arcpy.ListFields(table) if f.type!="Geometry"] 
summer_d2h_se_sq_wtd = pd.DataFrame(data=arcpy.da.SearchCursor(table, columns), columns=columns)

table = r"C:\Users\User\Documents\UNR\NEON Project\Data\GIS\WaterIsotopesDatabase\tables.gdb\winter_d2h_se_sq_wtd"
columns = [f.name for f in arcpy.ListFields(table) if f.type!="Geometry"] 
winter_d2h_se_sq_wtd = pd.DataFrame(data=arcpy.da.SearchCursor(table, columns), columns=columns)

#Elevation data
table = r"C:\Users\User\Documents\UNR\NEON Project\Data\GIS\WaterIsotopesDatabase\tables.gdb\DEM1km"
columns = [f.name for f in arcpy.ListFields(table) if f.type!="Geometry"] 
dem1km = pd.DataFrame(data=arcpy.da.SearchCursor(table, columns), columns=columns)

table = r"C:\Users\User\Documents\UNR\NEON Project\Data\GIS\WaterIsotopesDatabase\tables.gdb\DEM30m"
columns = [f.name for f in arcpy.ListFields(table) if f.type!="Geometry"] 
dem30m = pd.DataFrame(data=arcpy.da.SearchCursor(table, columns), columns=columns)


precip_d2h = {}
for s in range(len(sum_ppt_summer['SiteID'])):
    site = sum_ppt_summer['SiteID'].iloc[s]
    precip_d2h[s] = (site,
                     sum_ppt_x_d2h_summer.loc[sum_ppt_x_d2h_summer['SiteID'] == site,'SUM'].iloc[0] / 
                     sum_ppt_summer.loc[sum_ppt_summer['SiteID'] == site, 'SUM'].iloc[0],
                     sum_ppt_x_d2h_winter.loc[sum_ppt_x_d2h_winter['SiteID'] == site,'SUM'].iloc[0] / 
                     sum_ppt_winter.loc[sum_ppt_winter['SiteID'] == site, 'SUM'].iloc[0],
                     math.sqrt(summer_d2h_se_sq_wtd.loc[summer_d2h_se_sq_wtd['SiteID'] == site, 'SUM'].iloc[0] /
                     sum_ppt_summer.loc[sum_ppt_summer['SiteID'] == site, 'SUM'].iloc[0]),
                     math.sqrt(winter_d2h_se_sq_wtd.loc[winter_d2h_se_sq_wtd['SiteID'] == site, 'SUM'].iloc[0] /
                     sum_ppt_winter.loc[sum_ppt_winter['SiteID'] == site, 'SUM'].iloc[0]),
                     sum_ppt_summer.loc[sum_ppt_summer['SiteID'] == site, 'SUM'].iloc[0] /
                     sum_ppt_summer.loc[sum_ppt_summer['SiteID'] == site, 'COUNT'].iloc[0],
                     sum_ppt_winter.loc[sum_ppt_winter['SiteID'] == site, 'SUM'].iloc[0] /
                     sum_ppt_winter.loc[sum_ppt_winter['SiteID'] == site, 'COUNT'].iloc[0],
                     dem1km.loc[dem1km['SiteID'] == site, 'MEAN'].iloc[0],
                     dem30m.loc[dem30m['SiteID'] == site, 'MEAN'].iloc[0])
    
df = pd.DataFrame.from_dict(precip_d2h, orient='index', columns=['site', 'p_d2h_summer', 'p_d2h_winter', 'p_d2h_summer_se', 'p_d2h_winter_se', 'p_summer', 'p_winter', 'elev_1km', 'elev_30m'])

df.to_csv(r'C:\Users\User\Documents\UNR\NEON Project\Data\Compiled\precip_d2h_grid.csv')
```
